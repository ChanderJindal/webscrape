```python
import requests 
from bs4 import BeautifulSoup
```

```python
Base_Link = "https://www.readdetectiveconanarc.com/"
```

```python
Status = requests.get(Base_Link)
Status
```

```python
Page_html = Status.text
Page_html
```

```python

soup = BeautifulSoup(Page_html,features="lxml")
soup 
```

```python
Link = soup.find('a', class_ = "column has-text-centered button-last-chapter")
print(Link["href"],type(Link["href"]))
Link = str(Link["href"])
Link = Link[0:len(Link)-1]
print(Link)
```

```python
MangaEdition = int(Link.split('-')[-1])
MangaEdition
```
